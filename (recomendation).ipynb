```python
import pandas as pd
import numpy as np
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise import Reader
from surprise.model_selection import train_test_split
```

```python
ratings=pd.read_csv("C:/Users/Administrator/Downloads/rating.csv")
ratings.head()
```

```python
reader=Reader(rating_scale=(1,10))
data=Dataset.load_from_df(ratings[['User ID','Movie ID','Rating']],reader)
```

```python
[trainset,testset]=train_test_split(data,test_size=.15,shuffle=True)
```

```python
recom=KNNWithMeans(k=50,sim_options={'name':'cosine','user_based':True})
recom.fit(data.build_full_trainset()) #using full tarining set and not doning
test_pred=recom.test(testset)
RMSE=accuracy.rmse(test_pred)
```

```python
'''predict() is used to compute the rating prediction for given user and item'''
pre=recom.predict(1,6)
print(" The rating value for item 6 by user 1 is ",pre[3])

pre=recom.predict(2,2)
print(" The rating value for item 2 by user 2 is ",pre[3])

pre=recom.predict(2,4)
print(" The rating value for item 4 by user 2 is ",pre[3])

pre=recom.predict(3,1)
print(" The rating value for item 1 by user 3 is ",pre[3])

pre=recom.predict(3,6)
print(" The rating value for item 6 by user 3 is ",pre[3])

pre=recom.predict(4,5)
print(" The rating value for item 5 by user 4 is ",pre[3])
```

```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
store_data=pd.read_csv("C:/Users/Administrator/Downloads/store_data.csv",header=None)
store_data.shape
```

```python
records=[]
for i in range(0,7501):
    records.append([str(store_data.values[i,j])
                    for j in range(0,20) 
                    if (str(store_data.values[i,j])!='nan')])
    
    
    
'''
for i in range(o,7501):
     ls=[]
     for j in range(0,20):
        if (str(stroe_data.values[i,j]))
     records.append(ls)   '''    
```

```python
association_rules=apriori(records,min_support=.0045,min_confidence=.2,min_lift=3,max_length=3)
association_results=list(association_rules)
print(len(association_results))
```

```python
i=1
for item in association_results:
    #first index of iner list
    #contains base item and add items
    pair=item[0]
    items=[x for x in pair]
    print(i,"\n Rule: " + items[0] + "->" + items[1])
    
    #second index of the inner list
    print("Support: "+ str(item[1]))
    
    #third index of the list located as 0th
    #of the third index of the inner list
    
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
    i=i+1
```

```python
df=pd.DataFrame(columns=('Items','Antecedent','Consequent','Support','Confidence','Lift'))
Support=[]
Confidence=[]
Lift=[]
Items=[]
Antecedent=[]
Consequent=[]

for  RelationRecord in association_results:
    for ordered_stat in RelationRecord.ordered_statistics:
        Support.append(RelationRecord.support)
        Items.append(RelationRecord.items)
        Antecedent.append(ordered_stat.items_base)
        Consequent.append(ordered_stat.items_add)
        Confidence.append(ordered_stat.confidence)
        Lift.append(ordered_stat.lift)
        
        
df['Items']=list(map(set,Items))        
df['Antecedent']=list(map(set,Antecedent))
df['Consequent']=list(map(set,Consequent))
df['Support']=Support
df['Confidence']=Confidence
df['Lift']=Lift
df
```
